```python
from pymongo import MongoClient
from pprint import pprint
import pandas as pd

# Mengatur opsi tampilan pandas
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

```python
# Koneksi ke MongoDB (ubah parameter connection sesuai kebutuhan Anda)
client = MongoClient("mongodb://localhost:27017/")
db = client.dskola

# Fungsi untuk mengubah cursor MongoDB menjadi DataFrame pandas
def cursor_to_dataframe(cursor):
    return pd.DataFrame(list(cursor))
```

```python
# Task 1: Temukan semua buku yang ditulis oleh "Author 5" yang telah dipinjam oleh pengguna tetapi belum dikembalikan
task_1_result = db.books.find({
    "author": "Author 5",
    "borrowed_by.return_date": None
})
task_1_df = cursor_to_dataframe(task_1_result)
print("Task 1 Results:")
display(task_1_df)
```

```python
# Task 2: Daftar semua buku yang diterbitkan sebelum tahun 1980 dan memiliki lebih dari 5 salinan yang tersedia
task_2_result = db.books.find({
    "published_year": { "$lt": 1980 },
    "copies_available": { "$gt": 5 }
})
task_2_df = cursor_to_dataframe(task_2_result)
print("\nTask 2 Results:")
display(task_2_df)
```

```python
# Task 3: Temukan 5 buku terbaru yang diterbitkan dalam genre "Fantasy"
task_3_result = db.books.find({ "genre": "Fantasy" }).sort("published_year", -1).limit(5)
task_3_df = cursor_to_dataframe(task_3_result)
print("\nTask 3 Results:")
display(task_3_df)
```

```python
# Task 4: Hitung jumlah buku yang tersedia untuk setiap genre
task_4_result = db.books.aggregate([
    { "$group": { "_id": "$genre", "total_copies": { "$sum": "$copies_available" } } }
])
task_4_df = cursor_to_dataframe(task_4_result)
print("\nTask 4 Results:")
display(task_4_df)
```

```python
# Task 5: Temukan semua buku yang belum pernah dipinjam oleh pengguna manapun
task_5_result = db.books.find({ "borrowed_by": { "$eq": [] } })
task_5_df = cursor_to_dataframe(task_5_result)
print("\nTask 5 Results:")
display(task_5_df)
```
